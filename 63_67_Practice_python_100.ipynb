63. Write a program that can create a new list from a given list where each
item in the new list is square of the item of the old list

```python
def show(lis1):
  lis = lis1[:]
  lis2 = []
  for i in lis:
    lis2.append(i**2)
  return lis2

lis1 = list(map(int, input("Enter the element: ").split()))
print(show(lis1))
```

64. Write a program that can reverse words of a given string.
Eg if the input is Hello how are you
Output should be you are how Hello

```python
def show(str1):
  lis = str1[::-1]
  return ' '.join(lis)

st1 = input('Enter the words: ').split()
print(show(st1))
```

65. Write a program that can count the number of words in a given string

```python
def show(str1):
  count=0
  for i in str1:
    count+=1
  return count

str1 = input('Enter the words: ').split()
print(show(str1))
```

66. Write a program to check if a list is in ascending order or not

```python
def check(lis1):
  lis2 = sorted(lis1)
  if lis1 == lis2:
    return 'sorted'
  return 'not sorted'

lis1 = list(map(int, input('Enter the element: ').split()))
print(check(lis1))
```

67. Create 2 lists from a given list where 1st list will contain all the odd
numbers from the original list and the 2nd one will contain all the even
numbers  

```python
def show(lis1):
  lis2 = []
  lis3 = []

  for i in lis1:
    if i%2 == 0:
      lis2.append(i)
    else:
      lis3.append(i)

  return f"Even : {lis2}\n Odd: {lis3}"


lis1 = list(map(int, input("Enter the element: ").split()))
print(show(lis1))
```
