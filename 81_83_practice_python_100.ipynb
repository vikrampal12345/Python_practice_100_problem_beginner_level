81. Write a program to swap the key value pair for max and min values
Eg if the dict is like this {‘a’:1,’b’:2,’c’:3}
Output should be {a:3,b:2,c:1}

```python
# right method


dic = {
    'a' : 1,
    'b' : 2,
    'c' : 3,
  }

del max
del min  # it overwrite the dic so delete it and then the clear output

max_val = max(dic.values())
min_val = min(dic.values())


new_dic = {}
for key, value in dic.items():
  if value == max_val:
    new_dic[key] = min_val

  elif value == min_val:
    new_dic[key] = max_val

  else:
    new_dic[key] = value

new_dic

```

82. Write a program to find histogram of a given set of numbers. Take bin
size from user. Print the result in the form of a dictionary.

```python
numbers = list(map(int, input("Enter the element: ").split()))
bin_size = int(input('Enter the bin size: '))

hist = {}

start = (min(numbers)//bin_size)*bin_size
end = max(numbers)

while start <= end:
  count = 0
  for num in numbers:
    if start <= num <= start+bin_size:
      count += 1

  hist[f"{start}-{start+bin_size-1}"] = count
  start+=bin_size

print(hist)
```

83. Write a function that accepts a string and returns the number of upper
case chars and lower case chars as a dictionary

```python
# right method

def show(str1):
  count_l = 0
  count_u = 0
  dic = {}
  for i in str1:
    if i.isupper():
      count_u+=1
    # if i == i.upper():  # don't write like this
    #   count_u+=1
    elif i.islower():
      count_l+=1
  dic['Upper'] = count_u
  dic['Lower'] = count_l

  return dic

str1 = input("Enter the string: ").strip()
show(str1)
```
