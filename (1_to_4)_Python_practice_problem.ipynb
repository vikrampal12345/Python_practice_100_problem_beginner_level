## 1. User will input (3ages).Find the oldest one

method-1 : My approach

```python
def oldAge(age1, age2, age3):
  if age1 > age2 and age1 > age3:
    return age1
  elif age2 > age1 and age2 > age3:
    return age2
  else:
    return age3

oldAge(23, 34, 12)

```

method-2 : Best way (pythonic way to solve)

```python
def oldAge(age1, age2, age3):
  return max(age1, age2, age3)

oldAge(23,54,12)
```

## 2. Write a program that will convert celsius value to fahrenheit

```python
def celcius_to_farh(celc): #Naming should be like this
  ferh = (celc*9/5) + 32
  return ferh

celcius_to_farh(37)
```

## 3.User will input (2numbers).Write a program to swap the numbers

method-1 : My approach


```python
def swap_number(num1, num2):
  temp=num1
  num1=num2
  num2=temp
  return num1 , num2

swap_number(4,5)
```

method-2 : Best way (pythonic way)

```python
def swap_number(num1, num2):
  num1, num2 = num2, num1
  return num1, num2

swap_number(3,4)
```

# 4.Write a program that will give you the sum of 3 digits

```python
def sum_of_digit(num):
  sum=0
  for i in range(len(str(num))):
    digit=num%10
    sum+=digit
    num//=10
  return sum       # return function is the out of the loop

print(sum_of_digit(532))

```

method-2 : Best way

```python
def sum_of_digit(num):
    total = 0
    while num > 0:
        total += num % 10
        num //= 10
    return total
sum_of_digit(54223)
```
