76. Write a program to perform matrix multiplication on 2 matrices

```python
def matrix_mul(row1,col1,row2,col2):
    print("Enter first matrix: ")
    lis1 = []
    for i in range(row1):
      row = []
      for j in range(col1):
        lis = int(input('Enter the matrix: '))
        row.append(lis)
      lis1.append(row)
    print('Enter second matrix: ')
    lis2 = []
    for i in range(row2):
      row = []
      for j in range(col2):
        lis = int(input('Enter the matrix: '))
        row.append(lis)
      lis2.append(row)

    lis3 = []
    if col1 != row2:
      return 'Matrix multiplication is not possible'

    for i in range(row1):
      row = []
      for j in range(col2):
        total=0
        for k in range(col1):

            total+= lis1[i][k] * lis2[k][j]
        row.append(total)
      lis3.append(row)
    return lis3

row1 = int(input("Enter the row of the first matrix: "))
col1 = int(input("Enter the column of the first matrix: "))
row2 = int(input("Enter the row of the second matrix: "))
col2 = int(input("Enter the column of the second matrix: "))

print(matrix_mul(row1, col1, row2, col2))
```
