```python
# method- 3

def simplify(num, den):
    hcf = 1

    for i in range(1, min(num, den) + 1):
        if num % i == 0 and den % i == 0:
            hcf = i

    return f"{num // hcf}/{den // hcf}"

num = int(input("Enter numerator: "))
den = int(input("Enter denominator: "))

print(simplify(num, den))
```

52. Extract username from a given email.  Eg if the email is nitish24singh@gmail.com then the username should be nitish24singh 

```python
def username(gmail):
    lis=[]
    for letter in gmail:
        if '@' == letter:
            break
        lis.append(letter)

    return  ''.join(lis)
gmail=input("Enter the gmail: ")
print(username(gmail))        
```

53. Count the frequency of a particular character in a provided string. Eg 
'hello how are you' is the string, the frequency of h in this string is 2. 

```python
def count1(str1, ch1):
    count = 0
    for i in str1:
        if i == ch1:
            count +=1

    return count
str1 = input('Enter the words: ')
ch1 = input('Find the who many times repeat: ')
print(count1(str1, ch1))            
```

54. Find the index position of a particular character in another string. 

```python
def show(str1, ch1):
    for i in range(len(str1)):
        if ch1 == str1[i]:
            return i
    return 'the character is not exist in the word!'    
str1 = input('Enter the word: ')
ch1 = input('Enter the character which you want to their index: ')
print(show(str1, ch1))
```
