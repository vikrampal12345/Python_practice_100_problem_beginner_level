31. Write a program to print all the unique combinations of 1,2,3 and 4

```python
from itertools import combinations

digits = '1234'

for r in range(1, len(digits) + 1):
  print(list(combinations(digits, r)))
```

```python
from itertools import permutations

digits = '1234'

result = [''.join(p)  for p in permutations(digits)]

print(result)
```

34. Print first 25 prime numbers

```python
# num = int(input("Enter the number: "))
def number():
  lis = []
  for num in range(2, 26):

    for i in range(2, num):
      if num%i == 0:
         break
    else:
        lis.append(num)
  return lis
number()
```

35. Print the first 20 numbers of a Fibonacci series

```python
def fab(num):
  if num == 0:
    return 0
  if num == 1:
    return 1
  return fab(num-1) + fab(num-2)

num = int(input("Enter the nuber : "))

for i in range(num+1):
  print(fab(i), end=',')
```
