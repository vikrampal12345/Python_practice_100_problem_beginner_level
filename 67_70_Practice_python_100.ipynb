68. Write a program to merge 2 list without using the + operator

```python
lis1 = [1,2,3,4,5]
lis2 = [3,4,5,6,7]
print(lis1+lis2)
```

```python
lis1 = [1,2,3,4,5]
lis2 = [3,4,5,6,7]

for i in lis2:
  lis1.append(i)

print(lis1)
```

```python
# method - 3
lis1 = [1, 2, 3, 4, 5]
lis2 = [3, 4, 5, 6, 7]

lis1.extend(lis2)

print(lis1)
```

69. Write a program to replace an item with a different item if found in the
list  

```python
lis1 = list(map(int, input("Enter the element: ").split()))

for i in range(len(lis1)):
  lis1[i] = int(input("Enter the number: "))

print(lis1)

```

```python
# method - 2
def replace(lis1):
    for i in range(len(lis1)):
        lis1[i] = int(input(f"Enter element {i+1}: "))
    return lis1

lis1 = list(map(int, input("Enter the elements: ").split()))

print(replace(lis1))
```

```python
lis1 = [1,2,3,4,4,5]
lis1[1] = 5
print(lis1)
```

70. Write a program that can convert a 2D list to 1D listWrite a program that
can print  

```python
lis2 = [[1,2,3,4,5], [7,8,9]]
lis3 = []
for i in lis2:
  for j in i:
    lis3.append(j)
print(lis3)
```

```python
# method - 2
lis = [[1, 2, 3],
       [4, 5, 6],
       [7, 8, 9]]

lis1 = [j for i in lis for j in i]

print(lis1)
```
