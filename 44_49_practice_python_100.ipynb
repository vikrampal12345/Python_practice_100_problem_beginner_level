44. Write a program to print the following pattern

1

1 2 1

1 2 3 2 1

1 2 3 4 3 2 1

1 2 3 4 5 4 3 2 1

```python
def pattern():
  for i in range(1,6):

    for j in range(1,i+1):
      print(j, end='')

    for j in range(i-1, 0, -1):
      print(j, end='')
    print()

pattern()
```

46. Write a program to calculate the sum of the following series till the
nth term
1/1! + 2/2! + 3/3! + 4/4! +…….+ n/n!
n will be provided by the user

```python
n = int(input('Enter the number: '))
num=1
total=0
for i in range(1,n+1):
  num = num *i
  total+=i/num
print(total)
print(num)
```

47. Write a Python Program to Find the Sum of the Series till the nth term:  
1 + x^2/2 + x^3/3 + … x^n/n
n will be provided by the user

```python
n=int(input('Enter the nth power: '))
x=int(input('Enter the sum of the number'))
total=1
for i in range(2,n+1):
  total+=(x**i)/i

print(total)
```

48. The natural logarithm can be approximated by the following series.
If x is input through the keyboard, write a program to calculate the
sum of the first seven terms of this series.

```python

x=float(input('Enter the number: '))
a = (x-1)/x
total=a
for i in range(2,8):
  total += 0.5*(a**i)

print(total)
```
