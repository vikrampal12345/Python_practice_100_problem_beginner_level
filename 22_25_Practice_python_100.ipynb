# 21. Write a menu driven program - 1.cm to ft  2.kl to miles  3.usd to inr  4.exit

```python
class Solution:
  def __init__ (self, key, num):
    self.key = key
    self.num = num



  def menu(self):
    while True:
      if self.key == 1:
        return  f"Feet = {self.num/30.48:.2f} ft"
      elif self.key == 2:
        return f"Miles = {self.num*0.621371:.2f} miles"
      elif self.key == 3:
        return f"INR = {self.num * 90:.2f} INR"
      elif self.key == 4:
        return 'Exit program'
      else:
        return 'Invalid Input!'

print("1. Enter for the cm to ft:")
print("2. Enter for the km to miles: ")
print("3. Enter for the usd to INR : ")
print("4. Exit")
key = int(input("Choose (1-4) : "))
if key != 4:
  num = float(input("Enter the number: " ))
else:
  num = 0
app = Solution(key, num)
print(app.menu())
```

## 22. Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

```python
# head
# leg - checken - 2, dog - 4

class Solution:
  def __init__ (self, head, leg):
    self.head = head
    self.leg = leg

  def count(self):
    d = (self.leg - 2*self.head) / 2
    h = (4*self.head - self.leg) /2
    return f"dog = {int(d)}, hen = {int(h)}"

head = int(input("Enter the number of head: "))
leg =  int(input("Enter the number of leg: "))

app = Solution(head, leg)
print(app.count())
```

24. Write a program to find the sum of first n numbers, where n will be
provided by the user. Eg if the user provides n=10 the output should be 55.

```python
class Solution:
  def __init__(self, num):
    self.num = num

  def sum_of(self):
    total = 0
    nums = self.num
    if nums < 0:
      return 'Invalid input'
    for digit in range(0,nums+1):
      total+=digit
    return total
num=int(input("Enter the natural number: "))

app=Solution(num)
print(app.sum_of())
```
