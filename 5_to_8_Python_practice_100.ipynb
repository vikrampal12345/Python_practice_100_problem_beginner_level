## 5.  Write a program that will reverse a four digit number.Also it checks whether the reverse is true.

```python
# method 1
num=1234
num_str=str(num)
lis=[]
for i in range(len(num_str)):      # write for n numbers
   digit=num%10
   lis.append(digit)
   num//=10

result = "".join(str(x) for x in lis)
if result == num_str:
  print(False)
else:
  print(True)
```

```python
# method 2
def reverse_num(num):
  num_str=str(num)
  lis=[]
  while num>0:
    digit=num%10
    lis.append(str(digit))
    num//=10
  result=int("".join(lis))

  if result == num:
    return False
  else:
    return True

print(reverse_num(1234))
reverse_num(1234)
```

```python
# method-3
def reverse_num(num):
  return int(str(num)[::-1])
print(reverse_num(1234))
```

## 7.Write a program that will tell whether the given year is a leap year or not.

```python
# method 1
def leap_year(num):
  if (num%4 ==0) or (num%400==0 and num%100 !=0):
    return 'leap year'
  else:
    return 'not leap year'
print(leap_year(2023))
```

```python
# method 2
def leap_year(num):
  return 'leap year' if (num%4==0 and num%100!=0) or num%400==0 else 'not leap year'

print(leap_year(2024))
```

## 8.Write a program to find the euclidean distance between two coordinates

```python
# method 1
from math import sqrt
def euc_distance(corr):
  if corr == 2:
    def corr_2(x1,y1,x2,y2):
      return sqrt((x2-x1)**2 + (y2-y1)**2)
    return corr_2
  elif corr == 3:
    def corr_3(x1,y1,z1,x2,y2,z2):
      return sqrt((x2-x1)**2 + (y2-y1)**2 + (z2-z1)**2)
    return corr_3

distance_2 = euc_distance(2)
print(distance_2(2,4,1,2))
```

```python
# method - 2
from math import sqrt

def euc_distance(*corr):
  mid = len(corr) // 2
  p1 = corr[:mid]
  p2 = corr[mid:]
  return sqrt(sum((a-b)**2 for a, b in zip(p1,p2)) ) # zip make pair
print(euc_distance(1,2,3,4,5,6))   # 3D
print(euc_distance(2,3,4,6))   # 2D
```

# zip() Function

Definition:
zip() combines elements from two or more iterables position-wise into tuples.

Syntax:
zip(iterable1, iterable2, ...)

Visualization:

List 1 : [1, 2, 3]
List 2 : [A, B, C]

      zip()
         │
         ▼

[(1,A), (2,B), (3,C)]

Example:

names = ['Ram', 'Shyam']
marks = [85, 90]

list(zip(names, marks))

Output:
[('Ram', 85), ('Shyam', 90)]

Key Points:
• Combines elements index by index.
• Returns a zip object.
• Use list() to view the result.
• Stops when the shortest iterable ends.

Memory Trick:

List 1 ──┐
         ├── zip() ──► Paired Tuples
List 2 ──┘

Think: zip() works like a zipper that joins matching elements together.
