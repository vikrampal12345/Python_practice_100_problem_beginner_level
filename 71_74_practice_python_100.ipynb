71. Write a program that can perform union and intersection on 2 given list.  

```python
lis1 = list(map(int, input("Enter the element of list1: ").split()))
lis2 = list(map(int, input("Enter the element of list2: ").split()))
union = []
intersection = []

for i in lis1:
  if i not in union:
    union.append(i)

for i in lis2:
  if i not in union:
    union.append(i)

for i in lis1:
  if i in lis2:
    intersection.append(i)

print('Union', union)
print('intersection', intersection)
```

72. the max item of each row of a matrix.  

```python
lis = [[1,2,3,4],
       [4,4,2,1],
       [7,6,5,2]]


for i in lis:
  largest = i[0]

  for j in i:
    if j > largest:
      largest = j

  print(largest)



```

73. Write a program that can convert an integer to string.  

```python
num = 4545
str1 = str(num)
print(type(str1))
```

```python
# method - 2

num = int(input("Enter an integer: "))
temp = num
result = ""

while temp > 0:
  digit = temp % 10
  result = chr(digit+48) +result
  temp //=10
print(result)

print(type(result))
```
