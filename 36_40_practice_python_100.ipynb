36. Write a program to find the compound interest

```python
class Compound:
  def __init__(self, p, r, t):
    self.p = p
    self.r = r
    self.t = t

  def find(self):
    ci = self.p*(((1+self.r/100)**self.t)-1)
    return f"compound interest annual : {ci:.2f}"
p,r,t = map(float, input("Enter the principle, rate, time, resplectively:  ").split())

app = Compound(p,r,t)
print(app.find())
```

37. Write a Python program that accepts an integer (n) and computes the
value of n+nn+nnn.

```python
def cal(num):
  term=""
  total=0
  for n in range(3):
    term += str(num)
    total += int(term)

  return total

cal(5)
```

38. Take a number from the user and find the number of digits in it.

```python
def digit(num):
  total = 0
  for _ in range(len(str(num))):
    total+=1
  return total
digit(1234)
```

```python
def digit(num):
  count = 0

  while num != 0:
    num //=10
    count +=1
  return count

digit(1234)
```

```python
# method -2
def factor(num):
    factors = []

    for i in range(1, int(num ** 0.5) + 1):
        if num % i == 0:
            factors.append(i)

            if i != num // i:
                factors.append(num // i)

    return sorted(factors)

print(factor(36))
```
