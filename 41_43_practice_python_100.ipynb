41. Write a program to print the following pattern
*

**

'***'

'****'

'*****'

```python
def pattern():
  for i in range(6):
    for j in range(i):
      print('*', end='' )
    print()

pattern()
```

43. Write  a program to print the following pattern

      '*'

  '* * *'

  '* * * * *'

 '* * * * * * *'

'* * * * * * * * *'

```python
def pattern():
  n=9
  for i in range(1,n+1):

      for j in range(n-i):
         print(' ', end='')
      for j in range(i):
         print("*" , end=' ')

      print()

pattern()
```

```python
def pattern():
    n = 9

    for i in range(1, n + 1):

        for j in range(n - i):
            print(' ', end=' ')

        for j in range(i):
            print('*', end=' ')

        print()

pattern()
```

44. Write a program to print the following pattern
