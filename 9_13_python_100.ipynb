## 9. Write a program that take a user inputr of three angles and will find out whether it can form a triangle or not.

```python
# method - 1
def traingle(num1, num2, num3):
  if (num1+num2) > num3 and (num2+num3) > num1 and (num3+num1) > num2:
    return 'it is traingle'
  else:
    return 'it is not traingle'

num1=int(input("Enter the side of the traigle a: "))
num2=int(input("Enter the side of the traigle b: "))
num3=int(input("Enter the side of the traigle c: "))
print(traingle(num1, num2, num3))
```

```python
# method - 2
def traingle_angle(num1, num2, num3):
  if (num1 + num2 + num3 == 180) and num1 > 0 and num2 > 0 and num3 > 0 :
    return 'it is traingle'
  else:
    return 'it is not traingle'

num1 = int(input("Enter the angle of the traigle alpha: "))
num2 = int(input("Enter the angle of the traigle beta: "))
num3 = int(input("Enter the angle of the traigle gamma: "))
print(traingle_angle(num1, num2, num3))
```

```python
# method - 3
class Triangle:
  def __init__(self, a,b,c):
    self.angles = [a,b,c]

  def is_triangle(self):
    if sum(self.angles) == 180 and min(self.angles) >0:
      return 'it is triangle'
    return 'it is not triangle'

num1 = int(input("Enter the angle of the traigle alpha: "))
num2 = int(input("Enter the angle of the traigle beta: "))
num3 = int(input("Enter the angle of the traigle gamma: "))

t=Triangle(num1,num2,num3)
print(t.is_triangle())


```

## 10.Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit

```python
class Profit:
  def __init__(self, sp, cp):
    self.sp = sp
    self.cp = cp

  def find_p_l(self):
    if self.sp > self.cp:
      return f"profit = {self.sp - self.cp} and the profit % = {((self.sp - self.cp)/self.cp)*100:.2f}%"
    elif self.cp > self.sp:
      return f"loss = {self.cp - self.sp} and the loss % = {((self.cp - self.sp)/self.cp)*104:.2f}%"
    elif self.cp == self.sp:
      return 'no loss and no profit'
    else:
      return 'warning! you put wrong input, please check and retry5'
sp = int(input("Enter the selling price : "))
cp = int(input("Enter the cost price: "))

p = Profit(sp,cp)
print(p.find_p_l())
```

## 12. Write a program to find the volume of the cylinder. Also find the cost when ,when the cost of 1litre milk is 40Rs.

```python
# find - volumn of cylinder
# find - cost
from math import pi

class Cylinder:
  def __init__ (self, radius, height):
    self.radius = radius
    self.height = height

  def volume(self):

    if self.radius <= 0 or self.height <= 0 :
      return 'Invalid input, please check your input and try again'

    volume = pi*(self.radius**2)*self.height
    cost = volume * 40
    return (f"volume : {volume:.2f}\n"
            f"cost of the milk : {cost:.2f}")
r = int(input("Enter the radius of the cylinder: "))
h = int(input("Enter the height of the cylinder: "))
cl = Cylinder(r, h)
print(cl.volume())
```
