56. Write a program which can remove a particular character from a string.

```python
# method-1 my thinking
def rem(s, key):
  s1 = s
  for i in s1:
    if i == key:
      pass
    else:
      print(i, end='')

s = input('Enter the string: ').strip()
key = input("Enter the character which you want to remove: ")
rem(s,key)
```

```python
# method -2
def rem(s, key):
  result = ""
  for ch in s:
    if ch != key:
      result +=ch
  return result


s = input('Enter the string: ').strip()
key = input("Enter the character which you want to remove: ")
rem(s,key)
```

```python
# method-3
def rem(s, key):
  s = s.replace(key, "")
  return s

s = input('Enter the string: ').strip()
key = input("Enter the character which you want to remove: ")
rem(s,key)
```

57. Write a program that can check whether a given string is palindrome or
not.

```python
def palindrome(s):
  s = s.replace(" " , "").lower()
  s1 = s[::-1]
  if s1 == s:
    return 'palindrome'

  return 'Not palindrome'

s = input("Enter the string: ").strip()
print(palindrome(s))
```

59. Write a python program to convert a string to title case without using
the title()

```python
def show(s):
   s1 = s.split()
   result = ''
   for ch in s1:
    result += ch[0].upper() + ch[1:].lower() + " "
   return result.strip()
s = input("Enter the string : ").strip()
print(show(s))
```

60. Write a python program to find the max item from a list without using
the max function

```python
lis = [23, 1, 4, 6, 9, 0, 55, 87]
max = lis[0]
for i in lis:
  if i > max:
    max = i
print(max)

```

```python
# method - 5
lis = [23, 1, 4, 6, 9, 0, 55, 87]

left = 0
right = len(lis) - 1

while left < right:
    lis[left], lis[right] = lis[right], lis[left]
    left += 1
    right -= 1

print(lis)
```

62. Write a python program to search a given number from a list

```python
def find(lis, key):
  for i in range(len(lis)):
    if lis[i] == key:
      return i

    return 'Not found'
lis = list(map(int, input('Enter the numbers: ').split()))
key = int(input('Enter the key number: '))

print(find(lis, key))
```
