# 16. Write a program that will determine weather when the value of temperature and humidity is provided by the user.


```python
class Solution:
  def __init__(self, temp, humid):
    self.temp = temp
    self.humid = humid

  def temperature(self):
    if self.temp >= 30 and self.humid >=90:
      return 'Hot and Humid'
    elif self.temp >= 30 and self.humid < 90:
      return 'Hot'
    elif self.temp < 30 and self.humid >= 90:
      return 'Cool and Humid'
    elif self.temp < 30 and self.humid < 90:
      return 'Cool'

temp=int(input("Enter the temperture: "))
humid = int(input("Enter the Humid: "))
app = Solution(temp, humid)
print(app.temperature())
```

# 17. Write a program that will take three digits from the user and add the square of each digit

```python
def add_num(*num):
  if len(num) == 3:
    return sum(n**2 for n in num)
  else:
    return("Invalid input!, Please input only three numbers: ")

num = list(map(int, input("Enter the three number: " ).split()))
print(add_num(*num))

# the split function is store the value in list but
# when the map is used for the convert the string to int
# then it is map not list so we have to convert it into the list
```

# 18. Write a program that will check whether the number is armstrong number or not.

```python
class Solution:
  def __init__(self, num):
    self.num = num

  def armstrong(self):
    nums = self.num
    original = nums
    total = 0
    power = len(str(nums))
    for num in range(power):
      digit = nums%10
      total += (digit**power)
      nums//=10
    if total == original:
      return 'Armstrong number'
    return 'Not Armstrong number'
num = int(input("Enter the number: "))
app=Solution(num)
print(app.armstrong())
```

## Question 20: In-Hand Salary Calculator

Write a Python program to calculate an employee's **in-hand annual salary** after applying the required deductions.

### Requirements

The program should accept the employee's **annual gross salary** as input and apply the following deductions:

#### Fixed Deductions

* **HRA (House Rent Allowance):** 10% of the gross salary
* **DA (Dearness Allowance):** 5% of the gross salary
* **PF (Provident Fund):** 3% of the gross salary

#### Income Tax Slabs

| Annual Gross Salary     | Tax Rate    |
| ----------------------- | ----------- |
| ₹0 – ₹1,00,000          | Print `"K"` |
| ₹1,00,001 – ₹5,00,000   | 0%          |
| ₹5,00,001 – ₹10,00,000  | 10%         |
| ₹10,00,001 – ₹20,00,000 | 20%         |
| Above ₹20,00,000        | 30%         |

### Output

The program should display the following details:

1. Gross salary
2. HRA deduction amount
3. DA deduction amount
4. PF deduction amount
5. Tax amount
6. Final in-hand salary after all deductions

### Note

* Use conditional statements to determine the applicable tax slab.
* Do not use any predefined salary calculation libraries.
* Assume all deductions are calculated based on the annual gross salary.


```python
class Salary:
  def __init__ (self, salary):
    self.salary = salary

  def in_hand_salary(self):
    gross_salary = self.salary
    hra_duc = gross_salary*0.1
    dear_all = gross_salary*0.05
    pro_fund = gross_salary*0.03
    final_salary = gross_salary - hra_duc - dear_all - pro_fund

    if gross_salary < 0:
      return 'Invalid input!'
    elif 0 < gross_salary <= 100000:
      return (f"Gross Salary {gross_salary}\n"
              f"HRA(Hours Rent Allowance) : {hra_duc}\n"
              f"DA(Dearness Allowance) : {dear_all}\n"
              f"PF(Provident Fund) : {pro_fund}\n"
              f"Income Tax: K\n"
              f"Final In Hand Salary : {final_salary}\n")
    elif 100001 <= gross_salary <= 500000:
      return (f"Gross Salary {gross_salary}\n"
              f"HRA(Hours Rent Allowance) : {hra_duc}\n"
              f"DA(Dearness Allowance) : {dear_all}\n"
              f"PF(Provident Fund) : {pro_fund}\n"
              f"Income Tax : 0%\n"
              f"Final In Hand Salary : {final_salary}\n")
    elif 500001 < gross_salary <= 1000000:
      return (f"Gross Salary {gross_salary}\n"
              f"HRA(Hours Rent Allowance) : {hra_duc}\n"
              f"DA(Dearness Allowance) : {dear_all}\n"
              f"PF(Provident Fund) : {pro_fund}\n"
              f"Income Tax : {gross_salary*0.1}\n"
              f"Final In Hand Salary : {final_salary-(gross_salary*0.1)}")
    elif 1000001 <= gross_salary <= 2000000:
      return (f"Gross Salary {gross_salary}\n"
              f"HRA(Hours Rent Allowance) : {hra_duc}\n"
              f"DA(Dearness Allowance) : {dear_all}\n"
              f"PF(Provident Fund) : {pro_fund}\n"
              f"Income Tax : {gross_salary*0.2}\n"
              f"Final In Hand Salary : {final_salary-(gross_salary*0.2)}")
    elif 2000001 <= gross_salary:
      return (f"Gross Salary: {gross_salary}\n"
              f"HRA(Hours Rent Allowance) : {hra_duc}\n"
              f"DA(Dearness Allowance) : {dear_all}\n"
              f"PF(Provident Fund) : {pro_fund}\n"
              f"Income Tax : {gross_salary*0.3}\n"
              f"Final In Hand Salary : {final_salary-(gross_salary*0.3)}")

salary = int(input("Enter  your annual salary: "))
app = Salary(salary)
print(app.in_hand_salary())

```

```python
# method - 2
class Salary:
    def __init__(self, salary):
        self.salary = salary

    def in_hand_salary(self):
        gross_salary = self.salary

        if gross_salary < 0:
            return "Invalid input!"

        hra = gross_salary * 0.10
        da = gross_salary * 0.05
        pf = gross_salary * 0.03

        deductions = hra + da + pf

        if gross_salary <= 100000:
            tax = "K"
            tax_amount = 0

        elif gross_salary <= 500000:
            tax = "0%"
            tax_amount = 0

        elif gross_salary <= 1000000:
            tax = "10%"
            tax_amount = gross_salary * 0.10

        elif gross_salary <= 2000000:
            tax = "20%"
            tax_amount = gross_salary * 0.20

        else:
            tax = "30%"
            tax_amount = gross_salary * 0.30

        final_salary = gross_salary - deductions - tax_amount

        return (
            f"Gross Salary: {gross_salary}\n"
            f"HRA (House Rent Allowance): {hra}\n"
            f"DA (Dearness Allowance): {da}\n"
            f"PF (Provident Fund): {pf}\n"
            f"Income Tax: {tax}\n"
            f"Tax Amount: {tax_amount}\n"
            f"Final In-Hand Salary: {final_salary}"
        )


salary = int(input("Enter your annual salary: "))
app = Salary(salary)
print(app.in_hand_salary())
```
